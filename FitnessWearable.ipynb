```python
# function 1 (count the all equals values of a row)
def values_count(df, column):

    values = []
    labels = df[column].unique()
    for index in df[column].unique():
        values.append((df[column].value_counts()[index]))
    
    # you can take off this prints if you want to.
    # print('Labels: ', labels)
    # print('Repetition times: ', values)
        
    return(pd.DataFrame({'Labels':labels, 'Repetitions times':values}))


#function 2 (Do a dataframe based on two columns)

def something_per_something(df, column1, column2):
    dataframe = df[[column1, column2]]
    column1_list = (df[column1].unique())
    column2_list = (df[column2].unique())
    columns_transpose = []
    for item1 in column1_list:
        item_dataframe = dataframe[dataframe[column1] == item1]
        columns_transpose_lines = []
        for item2 in column2_list:
            columns_transpose_lines.append(item_dataframe[item_dataframe[column2] == item2].size / 2)
        columns_transpose.append(columns_transpose_lines)
    
    final_dataframe = pd.DataFrame()
    final_dataframe[column1] = column1_list
    final_columns = np.transpose(np.array(columns_transpose))
    
    x = 0
    for i in final_columns:
        final_dataframe[column2_list[x]] = i
        x += 1
        
    return final_dataframe


#function 3 (transforme a DataFrame of real numbers to one with percentage of the total number of a row)

def DataFrame_percentage(Original_Data_Frame, start_column, last_column):
    
    Original_Data_Frame['Total'] = Original_Data_Frame.sum(axis=1)
    Original_Data_Frame_percentage = Original_Data_Frame.copy()
    
    lista = []
    for column in Original_Data_Frame_percentage.iloc[:, start_column:last_column]:
         lista.append(Original_Data_Frame_percentage[column] * 100 / (Original_Data_Frame['Total']))
    a = 0
    for column in Original_Data_Frame_percentage.iloc[:, start_column:last_column]:
        Original_Data_Frame_percentage[column] = lista[a]
        a += 1
        Original_Data_Frame_percentage[column] = Original_Data_Frame_percentage[column]
    return Original_Data_Frame_percentage.drop('Total', axis=1)
```

```python
#Making the list of ages
Ages_list = DF_Ages['What is your age?'].unique()

Ages_list[1], Ages_list[0] = Ages_list[0], Ages_list[1] # swaping lines (I want under 18 before 18-24)

#Creating the list of people per age
People_per_age = [DF_Ages[DF_Ages['What is your age?'] == i ].size/2 for i in Ages_list]

#Creating the list of Female per age
Female_per_age = [
    DF_Ages[(DF_Ages['What is your age?'] == i) & (DF_Ages['What is your gender?'] == 'Female')].size/2 for i in Ages_list
]


#Creating the list of male per age
Male_per_age = [
    DF_Ages[(DF_Ages['What is your age?'] == i) & (DF_Ages['What is your gender?'] == 'Male')].size/2 for i in Ages_list
]

#Creating the list of 'other gender' per afe
Other_per_age = [
    DF_Ages[(DF_Ages['What is your age?'] == i) & (DF_Ages['What is your gender?'] == 'Prefer not to say')].size/2 for i in Ages_list
]

#Creating the dictionary to posteriorly make a DF
Dictionary_People_per_age = {
    'Ages':Ages_list,
    'Total people number': People_per_age,
    'Female': Female_per_age,
    'Male': Male_per_age,
    'Undefined': Other_per_age
}

#Doing the DF 'People per age and gender'
DF_People_per_age = pd.DataFrame(Dictionary_People_per_age)
display(DF_People_per_age)
```

```python
# Creating the graph of overall well-being age to show as exemple

WellBeing = something_per_something(GC, 'What is your age?', 'Do you feel that the fitness wearable has improved your overall well-being?')
display(WellBeing)

fig = px.line(WellBeing, x = 'What is your age?', y=['Neutral', 'Strongly agree', 'Agree', 'Disagree'])
fig.show()
```

This happens for two reasons. The first one is the big diference between the number of people of each age( We have way more younger people).the second one is the lack of data what increase the probability of overlap lines.

There is two ways to fix the first question. The first would be make a graph based not on the brute number of people but based on the percentage, something that i already done in others graphs. However the probability of overlap lines won't change, so we end up on the second way to do it: Atributte some value to each aswers sum all numbers from each age and than dividing for the total number of people from that age. Thereby doing a weighted average.

Values of each answer:

- Strongly disagree = -2
- Disagree = -1
- Neutral = 0
- Agree = 1
- Strongly Agree = 2

So we can see that as close the line gets to 2 means that the group agree more, and closer to -2 means that the people disagree more.

```python
# creating a total column of people to WellBeing_per_age dataframe

WellBeing_per_age = something_per_something(GC, 'What is your age?', 'Do you feel that the fitness wearable has improved your overall well-being?')
WellBeing_per_age['Total'] = WellBeing_per_age.iloc[:, [1,2,3,4]].sum(axis=1)

# swaping row under 18 and 18-24
WellBeing_per_age.iloc[1], WellBeing_per_age.iloc[0] = WellBeing_per_age.iloc[0].copy(), WellBeing_per_age.iloc[1].copy()

# Creating of weighted average column to WellBeing_per_age dataframe

for column in WellBeing_per_age.columns[1:-1]: # creating the columns with values of each type of answer
    if column == 'Strongly disagree': n = -2
    if column == 'Disagree': n = -1
    if column == 'Neutral': n = 0
    if column == 'Agree': n = 1
    if column == 'Strongly agree': n = 2
    WellBeing_per_age[f'{column} value'] = n * WellBeing_per_age[column]
display(WellBeing_per_age)



# creating the weighted average column
WellBeing_per_age['weighted average'] = WellBeing_per_age.iloc[:, [6,7,8,9]].sum(axis=1) / WellBeing_per_age['Total']
display(WellBeing_per_age)



# I am pretty sure that is other easier way to do this, that don't need the values columns like I did. I strongly recomend you
# to try do it by yourself.
```

```python
# creating the line graph of weighted average

fig = px.line(WellBeing_per_age, x = 'What is your age?', y=['weighted average'], title = 'Age by how much fitness improve your well being')
fig.update_yaxes(range=[-2, 2])
fig.show()
```
